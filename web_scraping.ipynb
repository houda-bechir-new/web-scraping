```python
#Importing necessary Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#website link
url='https://www.flipkart.com/search?q=apple%20mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
# Send a GET request to the URL
r=requests.get(url)
```

```python
# Parse the HTML content using BeautifulSoup
soup=BeautifulSoup(r.content,'html.parser')
#le commentaire suivant affiche le code HTML de la page contenant les données qu'on veux les collecter
#print(soup.prettify())
```

```python
container=soup.find_all('div', class_="_1AtVbE col-12-12")
print (container)
```

```python
#determine how many elements were found in the list
len(container)
```

```python
#lists
prices=[]
product=[]
ratings=[]
```

```python
# getting names
a=soup.find_all('div',class_='_4rR01T')
for n in a:
    ns=(n.text)
    product.append(ns)
```

```python
print(product)
```

```python
# getting price
price=soup.find_all('div', class_="_30jeq3 _1_WHN1")

for p in price:
    ps=(p.text)
    prices.append(ps)
```

```python
print(prices)
```

```python
#getting the ratings
rating=soup.find_all('div', class_="_3LWZlK")

for r in rating:
    rs=(r.text)
    ratings.append(rs)
```

```python
print(ratings)
```

```python
#create a Pandas DataFrame object from three lists: product, prices, and ratings
df=pd.DataFrame(list(zip(product,prices,ratings)), columns=["Product name","price","ratings"])
print(df)
```

```python
#convert dataframe to csv file
df.to_csv('apple.csv')
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.tutorialspoint.com/python/python_data_types.htm"
```

```python
#extract url html code using urllib
#read () : read content: output sequence of bytes
#decode; decode the sequence of bytes
html = urlopen(url).read().decode("utf-8") 
```

```python

#create a soup object from the given html code
# BeautifulSoup  parse the HTML content stored in the html variable.
#lxml: html parser libray
soup = BeautifulSoup(html, "lxml") 
print ( soup)
```

```python
#find and extract html code for the "Data Type Conversion Functions" table
table_code = soup.find_all(class_="table table-bordered")[1]
```

```python
#find and extract all table rows
rows = table_code.find_all("tr")[1:] # table data rows
```

```python

#find and extract data from each row
func_code = list()
funcs = list()
for row in rows:
    func_code.append(row)

for f in func_code:
    funcs.append(f.find_all("p"))
```

```python

#clean up data and remove html tags
for f in range(len(funcs)):
    for i in range(len(funcs[f])):
        funcs[f][i] = funcs[f][i].get_text()

```

```python
#turn the functions list into a pandas dataframe
df2 = pd.DataFrame(funcs, columns=["Function", "Description"])

print(df2)

```

```python
#convert dataframe to csv file
df2.to_csv('function.csv')
```
